# PCA

```python
def read_pc_file(pc_dat):
    time = []
    pc1 = []
    pc2 = []
    pc3 = []
    with open(pc_dat) as f:
        lines = f.readlines()
        for line in lines[1:]:
            _time, _pc1, _pc2, _pc3 = line.split()
            time.append(float(_time)/1000) # in us
            pc1.append(float(_pc1))
            pc2.append(float(_pc2))
            pc3.append(float(_pc3))
    return time, pc1, pc2, pc3
```

```python
time_run1, pc1_run1, pc2_run1, pc3_run1 = read_pc_file(pc_dat='projection_results.dat')
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)
xmin = 0
xmax = 10.1
ax.set_xlim([xmin, xmax])
ymin = -30
ymax = 51
ax.set_ylim([ymin, ymax])
plt.plot(time_run1[::1], pc1_run1[::1], color='black', label='PC1')
plt.plot(time_run1[::1], pc2_run1[::1], color='red',   label='PC2')
plt.plot(time_run1[::1], pc3_run1[::1], color='blue',  label='PC3')
plt.xlabel(r'Simulation time [$\mu s$]', fontsize=16)
plt.ylabel(r'Principal Component Scores', fontsize=16)
plt.xticks(np.arange(0, 10.1, 2))
plt.yticks(np.arange(-30, 51, 20))
ax.tick_params(labelsize=16)
plt.legend(loc='upper right', fontsize=20, frameon=False, ncol=3)
plt.show()
```
