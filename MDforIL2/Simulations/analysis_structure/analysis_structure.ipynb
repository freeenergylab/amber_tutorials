# RMSD

```python
def read_rmsd_file(rmsd_dat):
    time = []
    rmsd = []
    with open(rmsd_dat) as f:
        lines = f.readlines()
        for line in lines[1:]:
            _time, _rmsd = line.split()
            time.append(float(_time)*20/1000) # in ns
            rmsd.append(float(_rmsd))
    return time, rmsd
```

```python
time_run1, rmsd_run1 = read_rmsd_file(rmsd_dat='IL2_CA_rmsd.dat')
time_run2, rmsd_run2 = read_rmsd_file(rmsd_dat='SP4206_rmsd.dat')
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)
xmin = 0
xmax = 1001
ax.set_xlim([xmin, xmax])
ymin = 0
ymax = 15.
ax.set_ylim([ymin, ymax])
plt.plot(time_run2[::1], rmsd_run2[::1], color='blue',  label='SP4206') 
plt.plot(time_run1[::1], rmsd_run1[::1], color='black',  label='IL2') 
plt.xlabel('Simulation Time [ns]', fontsize=20)
plt.ylabel(r'RMSD [$\AA$]', fontsize=20)
plt.xticks(np.arange(0, 1001, 200))
plt.yticks(np.arange(0, 16, 5))
ax.tick_params(labelsize=16)
plt.legend(loc='upper right', fontsize=20, frameon=False, ncol=3)
plt.show()
```

# RMSF

```python
def read_rmsf_file(rmsf_dat):
    residue = []
    rmsf = []
    with open(rmsf_dat) as f:
        lines = f.readlines()
        for line in lines[1:]:
            _residue, _rmsf = line.split()
            residue.append(float(_residue))
            rmsf.append(float(_rmsf))
    return residue, rmsf
```

```python
residue_run1, rmsf_run1 = read_rmsf_file(rmsf_dat='rmsf_backbone.dat')
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)
xmin = 0
xmax = 131
ax.set_xlim([xmin, xmax])
ymin = 0
ymax = 9
ax.set_ylim([ymin, ymax])
plt.plot(residue_run1[::1], rmsf_run1[::1], color='blue', label='IL2') 
plt.scatter(residue_run1[::1], rmsf_run1[::1], color='blue') 
plt.xlabel('Residue ID', fontsize=20)
plt.ylabel(r'RMSF [$\AA$]', fontsize=20)
plt.xticks(np.arange(0, 131, 10))
plt.yticks(np.arange(0, 9, 2))
ax.tick_params(labelsize=16)
plt.legend(loc='upper right', fontsize=20, frameon=False, ncol=3)
plt.show()
```

# Distance between both ended C$\alpha$

```python
def read_dist_file(dist_dat):
    time = []
    dist = []
    with open(dist_dat) as f:
        lines = f.readlines()
        for line in lines[1:]:
            _time, _dist = line.split()
            time.append(float(_time)*20/1000) # in ns
            dist.append(float(_dist))
    return time, dist
```

```python
time_run1, dist_run1 = read_dist_file(dist_dat='end-to-end_dist.dat')
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)
xmin = 0
xmax = 1001
ax.set_xlim([xmin, xmax])
ymin = 0
ymax = 31.
ax.set_ylim([ymin, ymax])
plt.plot(time_run1[::1], dist_run1[::1], color='blue', label=r'Both terminal C$_\alpha$') 
plt.xlabel('Simulation Time [ns]', fontsize=20)
plt.ylabel(r'Distance [$\AA$]', fontsize=20)
plt.xticks(np.arange(0, 1001, 200))
plt.yticks(np.arange(0, 31, 5))
ax.tick_params(labelsize=16)
plt.legend(loc='upper right', fontsize=20, frameon=False, ncol=3)
plt.show()
```
