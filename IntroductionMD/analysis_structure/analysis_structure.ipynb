# RMSD

```python
def read_rmsd_file(rmsd_dat):
    time = []
    rmsd = []
    with open(rmsd_dat) as f:
        lines = f.readlines()
        for line in lines[1:]:
            _time, _rmsd = line.split()
            time.append(float(_time)*5/1000) # in ns
            rmsd.append(float(_rmsd))
    return time, rmsd
```

```python
time_run1, rmsd_run1 = read_rmsd_file(rmsd_dat='singleparticle_CA_rmsd.dat')
time_run2, rmsd_run2 = read_rmsd_file(rmsd_dat='single_CA_rmsd.dat')
time_run3, rmsd_run3 = read_rmsd_file(rmsd_dat='particle_CA_rmsd.dat')
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)
xmin = -0.01
xmax = 1.01
ax.set_xlim([xmin, xmax])
ymin = 0
ymax = 10.
ax.set_ylim([ymin, ymax])
plt.plot(time_run1[::1], rmsd_run1[::1], color='black',  label='SINGLEPARTICLE') 
plt.plot(time_run2[::1], rmsd_run2[::1], color='blue',  label='SINGLE') 
plt.plot(time_run3[::1], rmsd_run3[::1], color='red',  label='PARTICLE') 
plt.xlabel('Simulation Time (ns)', fontsize=20)
plt.ylabel(r'RMSD of C$\alpha$ ($\AA$)', fontsize=20)
plt.xticks(np.arange(0, 1.01, 0.2))
plt.yticks(np.arange(0, 10.1, 2))
ax.tick_params(labelsize=16)
plt.legend(loc='upper right', fontsize=13, frameon=False, ncol=3)
plt.show()
```

# RMSF

```python
def read_rmsf_file(rmsf_dat):
    residue = []
    rmsf = []
    with open(rmsf_dat) as f:
        lines = f.readlines()
        for line in lines[1:]:
            _residue, _rmsf = line.split()
            residue.append(float(_residue))
            rmsf.append(float(_rmsf))
    return residue, rmsf
```

```python
residue_run1, rmsf_run1 = read_rmsf_file(rmsf_dat='rmsf_backbone.dat')
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)
xmin = 0.9
xmax = 14.1
ax.set_xlim([xmin, xmax])
ymin = 0
ymax = 8.1
ax.set_ylim([ymin, ymax])
plt.plot(residue_run1[::1], rmsf_run1[::1], color='blue',  label='') 
plt.scatter(residue_run1[::1], rmsf_run1[::1], color='blue') 
plt.xlabel('Residue ID', fontsize=20)
plt.ylabel(r'RMSF ($\AA$)', fontsize=20)
plt.xticks(np.arange(1, 14.1, 1))
plt.yticks(np.arange(0, 8.1, 2))
ax.tick_params(labelsize=16)
# plt.legend(loc='upper right', fontsize=13, frameon=False, ncol=3)
plt.show()
```

# Distance between both ended C$\alpha$

```python
def read_dist_file(dist_dat):
    time = []
    dist = []
    with open(dist_dat) as f:
        lines = f.readlines()
        for line in lines[1:]:
            _time, _dist = line.split()
            time.append(float(_time)*5/1000) # in ns
            dist.append(float(_dist))
    return time, dist
```

```python
time_run1, dist_run1 = read_dist_file(dist_dat='end-to-end_dist.dat')
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)
xmin = -0.01
xmax = 1.01
ax.set_xlim([xmin, xmax])
ymin = 0
ymax = 50.
ax.set_ylim([ymin, ymax])
plt.plot(time_run1[::1], dist_run1[::1], color='blue',  label='') 
plt.xlabel('Simulation Time (ns)', fontsize=15)
plt.ylabel(r'Both ended C$\alpha$ distance ($\AA$)', fontsize=15)
plt.xticks(np.arange(0, 1.01, 0.2))
plt.yticks(np.arange(0, 50.1, 10))
ax.tick_params(labelsize=16)
# plt.legend(loc='upper right', fontsize=13, frameon=False, ncol=3)
plt.show()
```
